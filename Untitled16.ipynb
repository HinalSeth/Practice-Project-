# Variance Inflation Factor

```python
from scipy import stats
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
def calc_vif(x):
    vf = pd.DataFrame()
    vf['variables']=x.columns
    vf['VIF FACTOR']= [vif(x.values,i) for i in range(x.shape[1])]
    return vf
dfx1= df.drop(['primary_fuel','capacity_mw'],axis=1)
calc_vif(dfx1)
```

Since last 5 columns have infinite VIF value, and they all have same correlation with boththe target variabls and hence we will drop any four of them.

```python
dfx=dfx1.drop(['generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015', 'generation_gwh_2016'],axis=1)
calc_vif(dfx)
```

After dropping all the columns with multi-collinearity , we have VIF value in acceptable range.

# Model Building

# 1) Classification Models for predicting Primary Fuel

```python
x1= df.drop(['primary_fuel'],axis=1)
y1= df['primary_fuel']

from sklearn.preprocessing import power_transform
x1=power_transform(x1,method='yeo-johnson')

scale = StandardScaler()
x1=scale.fit_transform(x1)
```

We have separated the feature variables and the target variable, then we have treated our data for skewness and finally we have standardized our data.

```python
from imblearn.over_sampling import SMOTE
smt=SMOTE()
x1,y1 = smt.fit_resample(x1,y1)
sn.countplot(y1)
```

By using over_sampling we have balanced the classes.

```python
dt = DecisionTreeClassifier()
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=.30,random_state=45)
dt.fit(x_train,y_train)
dt.score(x_train,y_train)
preddt=dt.predict(x_test)

#from sklearn.naive_bayes import MultinomialNB
#mn=MultinomialNB()
#mn.fit(x_train,y_train)
#predmn=mn.predict(x_test)

rf = RandomForestClassifier()
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
predrf=rf.predict(x_test)

kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(x_train,y_train)
kn.score(x_train,y_train)
predkn=kn.predict(x_test)

sv = SVC()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)
predsv=sv.predict(x_test)

print("Accuracy of Decision Tree Classifier = ",accuracy_score(y_test,preddt)*100,"%")
#print("Accuracy of MultinomialNB = ",accuracy_score(y_test,predmn)*100,"%")
print("Accuracy of Random Forest Classifier =",accuracy_score(y_test,predrf)*100,"%")
print("Accuracy of K-Neighbor Classifier =",accuracy_score(y_test,predkn)*100,"%")
print("Accuracy of Support Vector Machine Classifier = ",accuracy_score(y_test,predsv)*100,"%")
```

We have printed accuracy of all the models together.

```python
print("CVS for DecisionTree Classifier =",cvs(dt,x1,y1,cv=5).mean()*100,"%")
#print("CVS for Logistic Regression =",cvs(lr,x,y,cv=3).mean()*100,"%")
print("CVS for Random Forest Classifier =",cvs(rf,x1,y1,cv=5).mean()*100,"%")
print("CVS for K-Neighbor Classifier =",cvs(kn,x1,y1,cv=5).mean()*100,"%")
print("CVS for Support Vector Machine Classifier =",cvs(sv,x1,y1,cv=3).mean()*100,"%")
```

Since decision tree classifier has minimum difference between accuracy and cvs , hence we will use this model for hyper parameter tuning.

```python
parameter = {'max_depth':np.arange(2,20),
            'criterion':['gini','entropy'],
            'splitter':['best','random'],
            'max_features':['auto','sqrt','log2']}
gvc = GridSearchCV(DecisionTreeClassifier(),parameter,cv=5)
gvc.fit(x_train,y_train)
gvc.best_params_
```

```python
Primary_fuel = DecisionTreeClassifier(criterion='entropy',max_depth=18,max_features='log2',splitter='best')
Primary_fuel.fit(x_train,y_train)
pred=Primary_fuel.predict(x_test)
acc=accuracy_score(y_test,pred)
print('Accuracy for predicting Primary Fuel is', (acc*100),'%')
```

# 2) Regression Model for predicting Capacity_mw

```python
x2= df.drop(['capacity_mw'],axis=1)
y2= df['capacity_mw']

from sklearn.preprocessing import power_transform
x2=power_transform(x2,method='yeo-johnson')

scale = StandardScaler()
x2=scale.fit_transform(x2)
```

We have separated the feature variables and the target variable, then we have treated our data for skewness and finally we have standardized our data.

```python
xtrain, xtest, ytrain, ytest = train_test_split(x2, y2, test_size = 0.2, random_state = 42)
lr= LinearRegression()
lr.fit(xtrain,ytrain)
lr.coef_
pred_train=lr.predict(xtrain)
pred_test=lr.predict(xtest)
print('r2_score of Linear Regression: ',r2_score(ytest,pred_test))

dtr=DecisionTreeRegressor(criterion='mse')
dtr.fit(xtrain,ytrain)
dtrpredict=dtr.predict(xtest)
print('r2_score of DecisionTree Regressor: ',r2_score(ytest,dtrpredict))

knr = KNeighborsRegressor()
knr.fit(xtrain,ytrain)
knrpredict=knr.predict(xtest)
print("r2_score of Kneighbors Regressor: ",r2_score(ytest,knrpredict))

svr = SVR()
svr.fit(xtrain,ytrain)
svrpredict=svr.predict(xtest)
print('r2_Score of SVR: ',r2_score(ytest,svrpredict))

sgd=SGDRegressor()
sgd.fit(xtrain,ytrain)
sgdpredict=sgd.predict(xtest)
print('r2_score of SGD: ',r2_score(ytest,sgdpredict))
```
